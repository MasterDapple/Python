```python
import serial
import time
import matplotlib.pyplot as plt
from tqdm import tqdm

# make sure the 'COM#' is set according the Windows Device Manager
ser = serial.Serial('COM3', 9800, timeout=1)
time.sleep(2)
t = 10 # Change this to increase the length of time to collect data, ie, 1 = 50 s, 2 = 100 s, etc

data = []
for i in range (t): 
    for q in tqdm(range(100), desc = "Loading..."): #This is just to prompt the loading bar
        line = ser.readline()
        if line: 
            bytes = line.decode('utf-8') #This decodes the input signals from unicode 
            num = float(bytes.replace("\r\n","")) #This allows for the string to be converted into a real
            data.append(num) #This appends the data to the existing list 

ser.close() #It's important to close the file you're accessing once you're done using it

#All of this essentially just makes the plot which allows us to visualize the data easier
plt.plot(data)
plt.xlabel('Time [s]')
plt.ylabel('Temperature Reading [C]')
plt.title('Temperature at Output of Thermal Box')
plt.show()
```
