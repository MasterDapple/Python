```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as interpolate
from scipy import linalg, optimize
import random


#This original function creates the plot trendline of the decoherence
def f_exact(x):
    return np.sin(x)/x

N = 10
x = np.linspace(0, 20, N)

y = f_exact(x)

fig, ax = plt.subplots()

x_fine = np.linspace(0, 20, 10*N)

ax.scatter(x, y)
ax.plot(x_fine, f_exact(x_fine), ls=":", label="original function")

#This part creates the randomized data which varies based on "noise"

#noise = random.uniform(-.5,.5)

def noisey(x):
    return x + random.uniform(-.1,.5)

fig, ax2 = plt.subplots()

ax2.scatter(x, y + noisey(y)) 
ax2.plot(x_fine, f_exact(x_fine), ls=":", label="original function")

```
