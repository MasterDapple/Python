```python
import pandas as pd
import csv 
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
```

```python
filename = 'QE.csv'
```

```python
fh = open(filename)
csvread = csv.reader(fh)
csvhead = next(csvread)
csvhead
```

```python
values = pd.read_csv(filename, header = None, skiprows = 2, names = csvhead)
values
```

```python
values.plot()
```

```python
gamma = values.CH4
charge = values.CH5
plt.plot(gamma, charge)
a, b = np.polyfit(gamma, charge,1)
plt.scatter(gamma,a*(gamma)+b)
plt.title("Quantum Efficiency")
plt.xlabel("walrus")
plt.ylabel("fetus")
plt.legend(["Measured Data","Trendline"])
```
